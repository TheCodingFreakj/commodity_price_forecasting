```python
import pandas as pd
data = pd.read_csv('commodity_price_data.csv')
data.head()
```

```python
data['MA_10'] = data['close_price'].rolling(window=10,  min_periods=1).mean()
data
```

```python
data['Momentum'] = data['close_price'].pct_change(periods=10)
data
```

```python
mean_value = data['close_price'].mean()
```

```python
data['Momentum'] = data['close_price'].fillna(mean_value)
data
```

```python
import numpy as np

delta = data['close_price'].diff(1)
gain = np.where(delta > 0, delta, 0)
loss = np.where(delta < 0, -delta, 0)

avg_gain = pd.Series(gain).rolling(window=14, min_periods=1).mean()
avg_loss = pd.Series(loss).rolling(window=14, min_periods=1).mean()
rs = avg_gain / avg_loss
data['RSI'] = 100 - (100 / (1 + rs))
data
clean_data = data.dropna()
clean_data
```

```python
# Calculate correlation matrix
correlation_matrix = clean_data.corr()

# Create the upper triangle of the correlation matrix (without the diagonal)
upper_triangle = correlation_matrix.where(
    np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)
)

# Find columns with correlation > 0.95 (but exclude 'close_price')
drop_columns = [column for column in upper_triangle.columns if any(upper_triangle[column] > 0.95) and column != 'close_price']

# Drop highly correlated columns (except 'close_price')
clean_data.drop(columns=drop_columns, inplace=True)

# Display the cleaned data
clean_data
```

```python
# Function to create sequences from DataFrame
def create_sequences(data, target, window_size=10):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data.iloc[i:i + window_size].values)  # Get the entire row (all columns) in each window
        y.append(target.iloc[i + window_size])  # Target is the next close_price from original data
    return np.array(X), np.array(y)

# Select the relevant columns (all columns except 'id' and 'close_price' as features)
data_features = clean_data.drop(columns=['id', 'close_price'])
# Target column 'close_price' from original data
target = clean_data['close_price']

# Create sequences from the DataFrame
X, y = create_sequences(data_features, target, window_size=10)

# Reshaping X for LSTM (samples, time_steps, features)
X = X.reshape((X.shape[0], X.shape[1], X.shape[2]))  # Now X is 3D (samples, time_steps, features)

# Check shapes
print("X shape:", X.shape)  # (samples, time_steps, features)
print("y shape:", y.shape)  # (samples, )
data_features
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
# Drop non-numeric columns (such as 'id', 'Month', 'Day_of_Week', etc.)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from scipy.stats import zscore
import numpy as np

# Assuming X_train and X_test are 3D arrays and the first column is a date column (index 0)
# X_train_numeric = X_train[:, :, 1:]  # Removing the date column
# X_test_numeric = X_test[:, :, 1:]  # Removing the date column
print(X_train.shape)
# Replace NaN or infinite values
X_train_clean = np.nan_to_num(X_train, nan=0, posinf=0, neginf=0)

# Ensure that the array is numeric
X_train_clean = np.array(X_train, dtype=np.float64)


X_train_reshaped = X_train.reshape(-1, X_train.shape[-1])  # Flattening to 2D (rows, features)

# Calculate z-scores for each feature
z_scores = zscore(X_train_reshaped, axis=0)

# Identify outliers (z-score > 3 or < -3)
outliers = np.abs(z_scores) > 2

# Print outliers detected
print(f"Outliers detected: \n{outliers}")


X_train_cleaned = X_train_reshaped[~np.any(outliers, axis=1)]
X_train_cleaned

# Calculate z-scores for each feature
z_scores = zscore(X_train_cleaned, axis=0)

# Identify outliers (z-score > 3 or < -3)
outliers = np.abs(z_scores) > 2

# Print outliers detected
print(f"Outliers detected: \n{outliers}")


# Parameters
timesteps = 10
num_features = 5
new_samples = 390  

# Total elements required
required_elements = new_samples * timesteps * num_features
print(required_elements)
print(X_train_cleaned.size)
print(X_train.size)
# Compute new number of samples after outlier removal
new_total_elements = X_train_cleaned.size
if new_total_elements % (timesteps * num_features) != 0:
    # Calculate the number of missing elements
    required_elements = (new_total_elements // (timesteps * num_features)) * (timesteps * num_features)
    remainder = new_total_elements % (timesteps * num_features)

    print(f"Data is incompatible for reshaping. Missing {remainder} elements.")
    print("Padding with zeros to make compatible...")

    # Pad with zeros
    padding_size = (timesteps * num_features) - remainder
    X_train_cleaned = np.pad(X_train_cleaned, (0, padding_size), mode='constant')

# Reshape back to 3D
num_samples = X_train_cleaned.size // (timesteps * num_features)
X_train_cleaned_3d = X_train_cleaned.reshape(num_samples, timesteps, num_features)

# Verify
print(f"Reshaped X_train_cleaned_3d shape: {X_train_cleaned_3d.shape}")
```

```python
from tensorflow.keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import RMSprop
from keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import tensorflow.keras.backend as K

# Define RMSE metric
def rmse(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true)))

# Initialize the model
model = Sequential()

# First LSTM Layer with BatchNormalization, Dropout, and L2 Regularization
model.add(LSTM(128,  # Increased LSTM units for better capacity
               input_shape=(X_train_scaled_3d.shape[1], X_train_scaled_3d.shape[2]),
               return_sequences=False,  # Set return_sequences to False for a simpler model
               kernel_regularizer=l2(0.001)))  # Regularization for better generalization
model.add(BatchNormalization())  # Normalize the activations of the LSTM layer
model.add(Dropout(0.5))  # Increased dropout rate for better regularization

# Output layer (for regression)
model.add(Dense(1, activation='linear'))  # No activation function for regression output

# Compile the model with RMSprop optimizer and loss function
optimizer = RMSprop(learning_rate=1e-5, clipvalue=1.0)  # Gradient clipping and reduced learning rate
model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

# Summary of the model
model.summary()
```
